```python
allfiles = [ file.name for file in scandir(folder) ]
```

```python
# Get the files for infant 1
files = [file for file in allfiles if str(1) in file]
```

```python
for i in range(1,8):
    files = [file for file in allfiles if str(i) in file]
    j=1
    for file in files:

        df = pd.read_csv(folder+file, usecols=col_dict[file])
        print(virtual_memory())

        plt.figure()
        for col in col_dict[file]:
            plt.plot(df[col])

        del df

        plt.legend(col_dict[file], loc="upper left")
        plt.title(file)
        plt.savefig("infant" + str(i) + "_part" + str(j) + ".jpg", dpi=300, bbox_inches="tight")
        plt.show()
        plt.close()
        gc.collect()
        print(virtual_memory())

```
