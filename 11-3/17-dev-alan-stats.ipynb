```python
infant = 1

df1 = pd.read_csv("rr_intervals/rr_ivls_" + str(infant) + ".csv")
df2 = pd.read_csv("troughless-rrivls/filtered_rpeaks_" + str(infant) + ".csv")
```

```python
print(df1.columns)
print(df2.columns)
```

```python
plt.figure()
plt.plot(df1["rpeaks"], df1["rr_ivl"])
plt.plot(df2["time"], df2["rrivl"])
plt.title("Infant " + str(infant) + " RR Intervals")
plt.xlabel("Interval placement (sec)")
plt.ylabel("Interval length (sec)")
plt.legend(["Unfiltered", "Filtered"], loc="upper left")
#plt.legend(["Filtered"], loc="upper left")
plt.show()
```

```python
# Okay, do I get the same thing if I just concatenate them here?
infant = 2

dfa = pd.read_csv("try_again2/rpeaks_" + str(infant) + "a.csv")
dfb = pd.read_csv("try_again2/rpeaks_" + str(infant) + "b.csv")
dfc = pd.read_csv("try_again2/rpeaks_" + str(infant) + "c.csv")
dfd = pd.read_csv("try_again2/rpeaks_" + str(infant) + "d.csv")
dfe = pd.read_csv("try_again2/rpeaks_" + str(infant) + "e.csv")
```

```python
plt.figure()
plt.plot(dfa["time"], dfa["time"].diff())
plt.plot(dfb["time"], dfb["time"].diff())
plt.plot(dfc["time"], dfc["time"].diff())
plt.plot(dfd["time"], dfd["time"].diff())
plt.plot(dfe["time"], dfe["time"].diff())
plt.title("Visibly Concatenated R-Peaks for Infant " + str(infant))
plt.xlabel("Interval Position (sec)")
plt.ylabel("Interval Duration (sec)")
```
